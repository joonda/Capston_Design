```python
gmaps_key = "AIzaSyD2s9NNwrCDki7vg_FT8IHlncLDgmEo1VA"
gmaps = googlemaps.Client(key=gmaps_key)

def get_location(location):
    result = gmaps.geocode(location, language="ko")
    lat = result[0].get("geometry")["location"]["lat"]
    lng = result[0].get("geometry")["location"]["lng"]
    formatted_address = result[0].get("formatted_address")

    return lat, lng, formatted_address

lat, lng, formatted_address = get_location("서울영등포경찰서")
```

```python
def get_route(start_lat, start_lng, end_lat, end_lng):

    url = "https://apis.openapi.sk.com/tmap/routes/pedestrian?version=1&callback=function"
    
    payload = {
        # 시작 지점 위/경도
        "startX": start_lng,
        "startY": start_lat,
        "angle": 20,
        "speed": 30,
        "endPoiId": "10001",
        # 종료 지점 (도착 지점) 위/경도
        "endX": end_lng,
        "endY": end_lat,
        #"passList": "126.92774822,37.55395475_126.92577620,37.55337145",
        "reqCoordType": "WGS84GEO",
        "startName": "%EC%B6%9C%EB%B0%9C", #출발이라는 한국어를 인코딩 시킨거 왜 했냐 한국어로 바꾸면 오류가 생길 수도 있다
        "endName": "%EB%8F%84%EC%B0%A9",
        "searchOption": "0",
        "resCoordType": "WGS84GEO",
        "sort": "index"
    }
    headers = {
        "accept": "application/json",
        "content-type": "application/json",
        "appKey": "e8wHh2tya84M88aReEpXCa5XTQf3xgo01aZG39k5"
    }
    
    response = requests.post(url, json=payload, headers=headers)
    
    print("URL:", url)
    print("QUERY PARAMS:")
    for key, value in payload.items():
        print(f"{key}: {value}")
    
    print("\nBODY PARAMS:")
    # Ensure proper encoding for Korean characters
    print(json.dumps(response.json(), indent=4, ensure_ascii=False).encode('utf-8').decode('utf-8'))
```

```python
start_location = '명지전문대학교'
goal_location = '홍대입구역'
```

```python
start_lat, start_lng, start_formatted_address = get_location(start_location)
end_lat, end_lng, end_formatted_address = get_location(goal_location)
```

```python
get_route(start_lat, start_lng, end_lat, end_lng)
```
